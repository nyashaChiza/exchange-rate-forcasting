Time Series Plot: This plot will give a clear understanding of how the exchange rate fluctuates over time. Since the data consists of monthly exchange rates spanning several years, a time series plot will allow us to visualize any long-term trends, seasonal patterns, and irregularities.

Moving Average Plot: Since we calculated moving averages for different windows (3-month, 6-month, and 12-month), plotting these alongside the original exchange rate can provide insights into the trend and smooth out short-term fluctuations, making it easier to identify long-term trends.

Seasonal Trend Plot: Since we added features like month_sin, month_cos, quarter_sin, and quarter_cos to capture seasonal patterns, plotting these can help visualize the cyclic behavior of the exchange rate over different months and quarters of the year. This plot can highlight any recurring patterns or seasonal effects.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path ='data/processed-data/feature_engineered_data.csv'
df = pd.read_csv(path)
```

```python
# Time Series Plot
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['rate'], label='Exchange Rate')
plt.title('Exchange Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Moving Average Plot
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['rate'], label='Exchange Rate')
plt.plot(df['date'], df['moving_avg_3'], label='3-Month Moving Average')
plt.plot(df['date'], df['moving_avg_6'], label='6-Month Moving Average')
plt.plot(df['date'], df['moving_avg_12'], label='12-Month Moving Average')
plt.title('Exchange Rate and Moving Averages')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.grid(True)
plt.legend()
plt.show()
```

The time series plot showcases the fluctuation of the exchange rate of the Zimbabwean dollar against the US dollar over time. Initially, the exchange rate remains relatively stable, hovering close to zero during the early period of observation. This stability is followed by a prolonged period where the exchange rate gradually increases, reaching approximately 500 units.

However, the most striking feature of the plot is the sudden spike in the exchange rate towards the end of the observed period, where it sharply rises to nearly 6000 units. This abrupt increase indicates a significant surge in the exchange rate, suggesting a period of extreme volatility and potentially adverse economic conditions.

Overall, the time series plot visually captures the dynamic nature of the exchange rate data, highlighting both long-term trends and abrupt changes in the exchange rate over time.

```python
# Seasonal Trend Plot
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['month_sin'], label='Month Sin')
plt.plot(df['date'], df['month_cos'], label='Month Cos')
plt.plot(df['date'], df['quarter_sin'], label='Quarter Sin')
plt.plot(df['date'], df['quarter_cos'], label='Quarter Cos')
plt.title('Seasonal Trends')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.show()
```
