checking for outliers in the data

```python
#using box plots
sns.boxplot(data=df)
```

The box plot visualizes the distribution of our ZWL (Zimbabwean Dollar) to USD exchange rate data. We observe that the majority of our data points are clustered around the lower end of the scale, with rates close to zero. This suggests that a significant portion of our data represents periods where the ZWL had also the same value compared to the USD.

However, we also notice that there is a wide range of values represented, with the upper end of the distribution extending up to around 1000 ZWL per USD. This indicates instances where the ZWL experienced significant fluctuations or periods of relative strength compared to the USD.

Additionally, the presence of outliers above the maximum whiskers is noteworthy. These outliers, reaching values as high as 6000 ZWL per USD, represent extreme instances of exchange rate volatility or irregularities that lie outside the typical range of our data. Further investigation may be warranted to understand the circumstances surrounding these outliers and their potential impact on our analysis.

```python
#using scatter plot
plt.scatter(df['date'], df['rate'])  # Replace 'x' and 'y' with the column names of the variables you want to plot
plt.xlabel('Date')
plt.ylabel('Rate')
plt.title('Scatter Plot')
plt.show()
```

The scatter plot illustrates the fluctuation of the ZWL (Zimbabwean Dollar) to USD exchange rate over the investigated period. Throughout most of the observed timeframe, the exchange rate demonstrates relatively consistent fluctuations, ranging from nearly 0 to approximately 3000 ZWL per USD. This suggests a significant degree of volatility in the currency exchange market during this period.

However, towards the end of the observed period, there is a notable departure from the typical fluctuation pattern. A sudden spike in the exchange rate is observed, with values surging to the 6000s ZWL per USD. This sharp increase indicates a period of exceptionally high volatility or instability in the currency market, potentially influenced by significant economic events or policy changes.

Overall, the scatter plot highlights the dynamic nature of the exchange rate, with periods of relatively stable fluctuations interspersed with episodes of pronounced volatility. Understanding the factors driving these fluctuations is crucial for comprehensively analyzing the economic dynamics of the region during the investigated period.
